# Tobacco Sales

This is an implementation of a time series linear regression for the quarterly prediction of tobacco sales in Spain. 

The data consists of:

  - Monthly total sales of tobacco in Spain.
  - Monthly sales of tobacco per province.
  - Number of tourist overnight stays per month.
  - Unemployment rate per quarter.


The data  consists of the total sales of tabacco in the whole in Spain, and also the total sales per province. Containing 50 provinces. The data also contains the total overnight stays, the unemployment rate and the price of tabacco. 

Using the prior monthly sales, number of tourist overnight stays and unemployent rate, we need to predict the total sale of tabacco per province and for total sales for Spain.

In order to do this two more linear regressions are implemented:

  - Linear regression for the prediction of overnight tourist stays.

  - Linear regression for the prediction of unemployment rate per quarter.

The data is then transformed to only contain quarterly information.

The data consist of datapoints from 2013 to 2016, we test our model predicting for 2017  

Beware data is written in spanish, the provinces also contain Spanish names, the code in itself and the graphs are all explained and written in English.

### 4.0 Sale prediction

Given the five data variables
  - Number of working days

  - Date
  - Tobacco average price
  
We will predict the quarterly price of tobacco.

```python
#Take concatenated data
df_ventas = pd.read_csv(path+"Datos ejercicio_predicted (1).csv",sep=';')
```

```python
#Drop unnecessary columns
df_ventas = df_ventas.drop([60,61,62])
```

```python
#Select variables
X = df_ventas[["Num. días laborables","Pernoctaciones","Tasa de paro","Average price","Fecha"]]
```

```python
X["Pernoctaciones"] = X["Pernoctaciones"].str.replace(",",".").astype(float)
X["Tasa de paro"] = X["Tasa de paro"].str.replace(",",".").astype(float)
X["Average price"] = X["Average price"].str.replace(",",".").astype(float)
```

```python
#Create Y by removing the X's variables
Y = df_ventas.drop(["Num. días laborables","Pernoctaciones","Tasa de paro","Average price","Fecha"],axis=1)
```

```python
#Cleanse names
for i in Y.columns:
  print(i)
  Y[i] = Y[i].str.replace(",",".")
  Y[i] = Y[i].astype(float)
```

#### 4.1 Preprocessing X

```python
#Modified date for label encoding
X["Fecha"] =X["Fecha"].str.replace("-","").str.replace("1","").str.replace("1","").str.replace("3","").str.replace("4","").str.replace("5","").str.replace("6","").str.replace("7","")
```

```python
#Apply one-hot encoding
encoding = pd.get_dummies(X["Fecha"])
```

```python
#Concat dataset with encoded dates
X = pd.concat([X,encoding],axis=1)
X.drop("Fecha",axis=1,inplace=True)
```

#### 4.2 Preprocessing

```python
#We apply train test split
y = Y.iloc[:48]
X = X.iloc[:48]
x_pred=  X.iloc[47:]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,shuffle=False)
```

#### 4.3 Model

```python
#Apply mode
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred_2016 = regressor.predict(X_test)  
#Check errors
y_pred_2017 = regressor.predict(x_pred)  
error_mse =mean_squared_error(y_test,y_pred_2016)
error_mae = mean_absolute_percentage_error(y_test,y_pred_2016)
print("Error mse de: ",error_mse)
print("Error mape de: ",error_mae)
```
